```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ptick
```

```python
plt.rcParams['font.size'] = 15
plt.rcParams['font.family'] = "serif"
fsize = 20
```

```python
data = np.loadtxt("./data2.csv",delimiter=",",unpack="True",skiprows=0)
nx,ny = np.shape(data)  #nx: number of column, ny: number of row
x = data[0,:]
y = np.array(data[1:nx,:])
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
mark = ["o","^","v","s","<",">","d","h","D","."]

for j in range(0,nx-1):
    ax.plot(x,y[j,:],mark[j],ms=10,label="Data No."+str(j+1))

ax.set_xlim(0,20)
ax.set_ylim(-0.5e-2,1e-2)

ax.yaxis.set_major_formatter(ptick.ScalarFormatter(useMathText=True))
ax.ticklabel_format(style="sci", scilimits=(0,0), axis="y")
ax.yaxis.offsetText.set_fontsize(18)
ax.tick_params(labelsize=18)

ax.legend(loc=0,fontsize=15,numpoints=1,ncol=2).get_frame().set_facecolor('#FFFFFF')
ax.set_xlabel(r"$x$",fontsize=fsize)
ax.set_ylabel(r"$f(x)$",fontsize=fsize)
ax.grid()
#fig.savefig("data2.png",dpi=300)
```
